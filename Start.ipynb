```
calendar = pd.read_csv(root+"/calendar.csv")
sell_prices = pd.read_csv(root+"/sell_prices.csv")
sales_train_validation = pd.read_csv(root+"/sales_train_validation.csv")
sample_submission = pd.read_csv(root+"/sample_submission.csv")
```

```
sell_prices['id'] = sell_prices['item_id']+'_'+sell_prices['store_id']+'_validation'
updated_sell = sell_prices.pivot(index='id', columns='wm_yr_wk', values='sell_price')
updated_sell = updated_sell.reindex(sales_train_validation['id'])
updated_sell = updated_sell.fillna(0)
```

```
old_column = updated_sell.columns
d =1
for i in range(len(old_column)):
  for j in range(7):
    updated_sell['d_{}'.format(d)]=updated_sell[old_column[i]]
    d+=1
  updated_sell.drop(old_column[i], axis=1, inplace=True)
```
