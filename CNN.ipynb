```
#Network Conv2D, MaxPooling2D, Dense, Flatten
def conv_net():
  x_in = Input(shape=(30490,32,2))
  x = Conv2D(2,(10,10),strides=(1,1), padding='same', activation='relu')(x_in)
  x = MaxPooling2D((3,3),padding='same')(x)
  # x = Conv2D(64,(10,10),strides=(1,1), padding='same', activation='relu')(x)
  # x = MaxPooling2D((3,3),padding='same')(x)
  x = Flatten()(x)
  #x = Dense(30490*2, activation='relu')(x)
  x_out = Dense(30490)(x)
  return tf.keras.Model([x_in],[x_out])
```
